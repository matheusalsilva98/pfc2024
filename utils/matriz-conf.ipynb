```python
import imageio.v2 as imageio
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gt_dir = '/media/reginaldo/pfc-dados/dataset-piloto/testes/treino1/02-11-21/mascara-real.tif'
img_gt = imageio.imread(gt_dir)
```

```python
treino1_dir = '/media/reginaldo/pfc-dados/dataset-piloto/testes/treino1/02-11-21/02-11-21-treino1.tif'
img_treino1 = imageio.imread(treino1_dir)

cm = confusion_matrix(np.ravel(img_gt), np.ravel(img_treino1))

# Converter para porcentagem
matriz_confusao_percent = cm / cm.sum(axis=1, keepdims=True) * 100

# visualize confusion matrix with sklearn metrics
# define matrix and create labels
cm_display = ConfusionMatrixDisplay(confusion_matrix = matriz_confusao_percent, display_labels = ['fundo', 'nuvem densa', 'nuvem fina', 'sombra'])

# display matrix
cm_display.plot()
plt.title('Matriz de Confus√£o para o Treino 1')
plt.show()
```
